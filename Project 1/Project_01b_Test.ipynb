## Test 01 

```python
np.random.seed(1)
X1a, y1a = make_circles(n_samples=500, noise=0.1, factor=0.7)
X1b, y1b = make_circles(n_samples=500, noise=0.15, factor=0.3)
X1 = np.vstack([X1a, X1b*0.5])
y1 = np.hstack([y1a, 2 - y1b])

plt.scatter(X1[:,0], X1[:,1], c=y1, edgecolor='k', cmap='rainbow')
plt.show()
```

```python
X1_train, X1_val, y1_train, y1_val = train_test_split(X1, y1, test_size=0.2, random_state=1)
```

```python
np.random.seed(1)
model_1 = ANN(X1_train, y1_train, layer_sizes=[2, 8, 16, 8, 3], activation='relu')

plot_regions(model_1, X1_train, y1_train)
```

```python
model_1.train(2000, lr=0.0001, display=100)

plot_regions(model_1, X1_train, y1_train)
```

```python
print('Training Accuracy:  ', model_1.score(X1_train, y1_train)[1])
print('Validation Accuracy:', model_1.score(X1_val, y1_val)[1])
```
